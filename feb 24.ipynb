```python
# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
# DataFrame with a new index that starts from 1 and increments by 2 for each row.
```

```python
# Ans 2
# Sure! You can use the set_index() method of the Pandas DataFrame to re-index the DataFrame with a new 
# index that starts from 1 and increments by 2 for each row. Here's an example Python function that does this:
import pandas as pd

def reindex_df(df):
    new_index = pd.RangeIndex(start=1, stop=2*len(df), step=2)
    return df.set_index(new_index)
# Here, we first create a new RangeIndex object that starts from 1, stops at 2 
# times the length of the original DataFrame, and increments by 2. We then use the set_index() 
# method of the DataFrame to set this new index.
# You can call this function by passing your DataFrame as an argument:

# python
new_df = reindex_df(df)
# This will return a new DataFrame with the desired index.
```

```python
# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
# iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
# function should print the sum to the console.
```

```python
# Ans 3=
import pandas as pd

def sum_first_three(df):
    total = 0
    for i, value in enumerate(df['Values']):
        if i < 3:
            total += value
        else:
            break
    print("Sum of first three values:", total)
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
sum_first_three(df)
```

```python
# Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
# 'Word_Count' that contains the number of words in each row of the 'Text' column.
import pandas as pd

def add_word_count(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(x.split()))
    return df
df = pd.DataFrame({'Text': ['This is the first sentence.', 'This is the second sentence.', 'This is the third sentence.']})
new_df = add_word_count(df)
```

```python
# Q5. How are DataFrame.size() and DataFrame.shape() different?
# Ans 5 =# DataFrame.size returns the total number of elements in the DataFrame, which is equal to the number 
# of rows multiplied by the number of columns. In other words, it returns the total number of cells in
# the DataFrame, including empty cells.
 # DataFrame.shape returns a tuple representing the dimensions of the DataFrame, with the number of rows 
# and columns as the first and second elements of the tuple, respectively. It does not include empty cells in its count.
# Here's an example to illustrate the difference:
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, None]})
print(df.size)    # 6
print(df.shape)   # (3, 2)
# In this example, the DataFrame df has 6 total elements (2 columns x 3 rows), including one 
# missing value. The shape of the DataFrame is (3, 2), indicating that it has 3 rows and 2 columns.
```

```python
Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.

The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.
Ans-
    To extract the username from each email address in a Pandas DataFrame df that contains an 'Email' column, you can use the str.split() method to split the email address into two parts: the username and the domain. Then, you can select the first part of the split using indexing to extract the username.

Here is a Python function that creates a new column 'Username' in the DataFrame df that contains only the username part of each email address:
    
    def extract_username(df):
    # Split the 'Email' column into two parts: the username and the domain
    df[['Username', 'Domain']] = df['Email'].str.split('@', expand=True)

    # Return the updated DataFrame
    return df

In this function, the str.split() method is used to split the 'Email' column of the DataFrame df into two parts: the username and the domain. The expand=True parameter is used to expand the split result into separate columns. The [['Username', 'Domain']] syntax is used to select these two columns and assign them to the DataFrame.

After the split, the 'Username' column contains the username part of each email address, and the 'Domain' column contains the domain part of each email address. You can then return the updated DataFrame df that includes the 'Username' and 'Domain' columns.

Here is an example of how to use this function:
    
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'Email': ['john.doe@example.com', 'jane.doe@example.com', 'james.smith@example.com']})

# Call the function to extract the usernames
df = extract_username(df)

# Display the updated DataFrame
print(df)

This would output:

                   Email Username        Domain
0  john.doe@example.com  john.doe    example.com
1  jane.doe@example.com  jane.doe    example.com
2  james.smith@example.com james.smith example.com


As you can see, the function has created a new 'Username' column in the DataFrame df that contains only the username part of each email address.
```

```python
Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
function should return a new DataFrame that contains only the selected rows.
For example, if df contains the following values:
A B C
0 3 5 1
1 8 2 7
2 6 9 4
3 2 3 5
4 9 1 2

Your function should select the following rows: A B C
1 8 2 7
4 9 1 2
The function should return a new DataFrame that contains only the selected rows.

Ans-

To select all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10 in a Pandas DataFrame df, you can use boolean indexing. Here is a Python function that performs this selection and returns a new DataFrame that contains only the selected rows:
    
def select_rows(df):
    # Use boolean indexing to select rows where 'A' > 5 and 'B' < 10
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]

    # Return the new DataFrame with the selected rows
    return selected_rows 
    
In this function, the [(df['A'] > 5) & (df['B'] < 10)] syntax is used to select only the rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10.

After the selection, the selected_rows DataFrame contains only the selected rows. You can then return this new DataFrame.

Here is an example of how to use this function: 
    
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': [3, 8, 6, 2, 9],
                   'B': [5, 2, 9, 3, 1],
                   'C': [1, 7, 4, 5, 2]})

# Call the function to select the rows
selected_rows = select_rows(df)

# Display the selected rows
print(selected_rows)

This would output:
    
A  B  C
1  8  2  7
4  9  1  2


As you can see, the function has selected only the rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10, and returned a new DataFrame that contains only these selected rows.    
```
